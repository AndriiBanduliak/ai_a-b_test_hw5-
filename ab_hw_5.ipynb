## Bы провели эксперимент c упрощением формы заказа в магазине Утконос и получили результаты по метрике конверсий в покупку. Выберите метод оценки и оцените есть ли стат.значимые различия между конверсиями в двух группах при alpha = 5%. Дайте краткие рекомендации команде .Результаты: 1)Число юзеров в группах , которые заходили на сайт в период эксперимента: n1 = 15550 и n2 = 15550 . 2) Число юзеров в группах , которые совершили хотя бы одну покупку за период эксперимента: n1 = 164 и n2 = 228 3) Конверсии : conv1 = 1.05% conv2 = 1.47% .

```python
#Оценка статистической значимости различий между конверсиями в двух группах с использованием Z-теста

import math

# размеры выборок
n1 = 15550
n2 = 15550

# число юзеров, совершивших покупку
m1 = 164
m2 = 228

# конверсии
conv1 = m1 / n1
conv2 = m2 / n2

# стандартные отклонения
std1 = math.sqrt(conv1 * (1 - conv1))
std2 = math.sqrt(conv2 * (1 - conv2))

# Z-статистика
z_score = (conv2 - conv1) / math.sqrt(std1 ** 2 / n1 + std2 ** 2 / n2)

# критическое значение
alpha = 0.05
critical_value = 1.96

# проверка гипотезы
if abs(z_score) > critical_value:
    print("Различия в конверсии между группами статистически значимы")
else:
    print("Различия в конверсии между группами не статистически значимы")
```

###  вычислили Z-статистику и сравнили ее с критическим значением для выбранного уровня значимости alpha. Различия между конверсиями в группах статистически значимы, так как Z-статистика превышает критическое значение.

## Если размер выборки в каждой группе составляет 10000 элементов, то для сравнения метрики конверсии в покупку можно использовать как Z-тест, так и t-тест Стьюдента.

```python
import math

# размеры выборок
n = 10000

# число юзеров, совершивших покупку
m1 = 164
m2 = 228

# конверсии
conv1 = m1 / n
conv2 = m2 / n

# стандартные отклонения
std1 = math.sqrt(conv1 * (1 - conv1))
std2 = math.sqrt(conv2 * (1 - conv2))

# Z-статистика
z_score = (conv2 - conv1) / math.sqrt(std1 ** 2 / n + std2 ** 2 / n)

# критическое значение
alpha = 0.05
critical_value = 1.96

# проверка гипотезы
if abs(z_score) > critical_value:
    print("Различия в конверсии между группами статистически значимы")
else:
    print("Различия в конверсии между группами не статистически значимы")
```

### вычислили Z-статистику и сравнили ее с критическим значением для выбранного уровня значимости alpha. Различия между конверсиями в группах статистически значимы, так как Z-статистика превышает критическое значение.

```python
# t-тест Стьюдента

from scipy import stats

# размеры выборок
n = 10000

# число юзеров, совершивших покупку
m1 = 164
m2 = 228

# конверсии
conv1 = m1 / n
conv2 = m2 / n

# t-статистика
t_score, p_value = stats.ttest_ind_from_stats(conv1, std1, n, conv2, std2, n)

# критическое значение
alpha = 0.05
critical_value = stats.t.ppf(1 - alpha / 2, 2 * n - 2)

# проверка гипотезы
if abs(t_score) > critical_value:
    print("Различия в конверсии между группами статистически значимы")
else:
    print("Различия в конверсии между группами не статистически значимы")
```
